```python
#productname=soup.find("div",attrs={"class":"RG5Slk"}) #pname is product name
#features=soup.find("ul",attrs={"class":"HwRTzP"})
#discount_percentage=soup.find("div",attrs={"class":"HQe8jr"})
#original_price=soup.find("div",attrs={"class":"kRYCnD gxR4EY"})
#discount_price=soup.find("div",attrs={"class":"hZ3P6w DeU9vF"})
#reviews=soup.find("span",attrs={"class":"PvbNMB"})
#ratings=soup.find("div",attrs={"class":"MKiFS6"})

from bs4 import BeautifulSoup
import requests
import numpy as np
import time

productname = [] 
rating = []
features = []
pagenum = []
original_price = []
discount_price= []
Review = []
discount_percentage=[]

total_time = time.time()

for i in range(1, 89): 
    start_time = time.time()
    URL = f"https://www.flipkart.com/search?q=air+conditioner&sid=j9e%2Cabm%2Cc54&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_8_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_8_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=air+conditioner%7CAir+Conditioners&requestId=fa896ed2-865c-4827-af01-1baa4e269306={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.text, "html.parser")
    for x in soup.find_all('div', attrs={'class':'lvJbLV col-12-12'}):
        pname= x.find('div', attrs={'class':'RG5Slk'})  #product name
        cost = x.find('div', attrs={'class':'hZ3P6w DeU9vF'})   # discounted price
        rat = x.find('div', attrs={'class':'MKiFS6'})           #rating
        feac = x.find('ul', attrs={'class':'HwRTzP'})                #features
        op = x.find('div', attrs={'class':'kRYCnD gxR4EY'})    #original price
        discount_pec= x.find('div', attrs={'class':'HQe8jr'}) #discount percentage
        reviews = x.find('span', attrs={'class':'PvbNMB'})      #reviews
        
        

        productname.append(pname.text if pname else np.nan)
        discount_price.append(cost.text if cost else np.nan)
        rating.append(rat.text if rat else np.nan)
        features.append(feac.text if feac else np.nan)
        original_price.append(op.text if op else np.nan)
        discount_percentage.append(discount_pec.text if discount_pec else np.nan)
        Review.append(reviews.text if reviews else np.nan)
        
        pagenum.append(i)

    print(f'Page {i} completed in {time.time()-start_time:.2f} seconds')

print("Total Time Completed in seconds", str(time.time()-total_time))
```

```python
# Create a dictionary first
df= {
    "Page Number": pagenum,
    "productname": productname,
    "discountprice": discount_price,
    "rating": rating,
    "features": features,
    "originalprice": original_price,
    "discountpercentage": discount_percentage,
    "reviews": Review,
}

# Convert the dictionary into a DataFrame
data= pd.DataFrame(df)
```

```python
outlier_summary = []

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outlier_count = ((df[col] < lower) | (df[col] > upper)).sum()
    
    outlier_summary.append({
        "Column Name": col,
        "Outlier Count": outlier_count
    })

outlier_df = pd.DataFrame(outlier_summary)
print(outlier_df)
```

```python
import pandas as pd
import numpy as np

# Load data
#df = pd.read_excel("output (2).xlsx")

cols = ['discountprice', 'originalprice', 'UnitsRoomSize', 'Reviews_Count']

# Convert to numeric safely
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Function to get outliers
def get_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    return data[(data[column] < lower) | (data[column] > upper)]

# Extract outliers
outliers = {}

for col in cols:
    outliers[col] = get_outliers(df, col)

# Print counts
for col, data in outliers.items():
    print(f"{col} outliers count:", data.shape[0])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


# ------------------------------
# 1️⃣ CLEAN THE PRICE COLUMN
# ------------------------------

df["originalprice"] = (
    df["originalprice"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

# ------------------------------
# 2️⃣ GET TOP 5 BRANDS (BY AVG PRICE)
# ------------------------------

top5 = (
    df.groupby("Brand")["originalprice"]
      .mean()
      .sort_values(ascending=False)
      .head(5)
)

# ------------------------------
# 3️⃣ BARPLOT
# ------------------------------

plt.figure(figsize=(12,7))
sns.barplot(
    x=top5.index,
    y=top5.values,
    palette="viridis"
)

plt.title("Top 5 Brands by Average Original Price", fontsize=18)
plt.xlabel("Brand", fontsize=14)
plt.ylabel("Average Original Price", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
